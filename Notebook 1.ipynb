### 1 Find the relationship between top 5 Job designation and total Purchase amount

```
Q1_df = df.head(5)
Q1_df
```

```
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 6))  # Optional: Set the figure size

sns.barplot(x='Job', y='Purchase Price', data=Q1_df)
plt.xlabel('Job')
plt.ylabel('Purchase Price')
plt.title('Relationship between top 5 Job designation and Purchase Price')

plt.show()
```

### 2 Find the relationship between Job designation and mean Purchase amount

```
df.groupby('Job')['Purchase Price'].mean()
```

```
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))  # Optional: Set the figure size

plt.plot(x='Job', y='Purchase Price', data=df.groupby('Job')['Purchase Price'].head(10).mean())
plt.xlabel('Categorical Column')
plt.ylabel('Mean of Column')
plt.title('Relationship between Categorical Column and Mean of Column')

plt.xticks(rotation=90)  # Optional: Rotate x-axis labels if needed

plt.show()
```
