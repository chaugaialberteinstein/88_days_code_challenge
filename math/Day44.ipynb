Expected result:


       val_1  val_2  entropy
    0   0.01   0.99   0.0808

```python
import pandas as pd
import numpy as np
```

```python
from google.colab import files 
uploaded = files.upload()
```

```python
import io
np.set_printoptions(precision=4, suppress=True)
df = pd.read_excel(io.BytesIO(uploaded['entropy.xlsx']))
```

```python
def entropy(x):
  return np.round(-np.sum(x * np.log2(x)), 4)
```

```python
df['entropy'] = [
    entropy([row[1][0], row[1][1]]) 
    for row in df.iterrows()
]
 
```
