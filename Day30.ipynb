```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
```

```python

lbc = load_breast_cancer()
data = lbc['data']
target = lbc['target']

all_data = np.c_[data,target]

df = pd.DataFrame(
    data=all_data, 
    columns=list(lbc['feature_names']) + ['target']
)
```
